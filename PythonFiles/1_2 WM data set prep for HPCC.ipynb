### RTMA only data

```python
RTMA_df = pd.read_csv('/Users/jilliancheck/OneDrive - Michigan State University/Documents/Work/White mold/Modeling/jupyter notebooks/WM train test data Fall 24/RTMA_23_df.csv')

# Create row shading 

row_shading = RTMA_df['canopy_avg'] * RTMA_df['spacing (m)']
insert_index = RTMA_df.columns.get_loc('target') + 1  # Insert after col1
RTMA_df.insert(insert_index, 'row_shading', row_shading)

# Remove or encode categorical data

RTMA_df = RTMA_df.drop(columns=['index', 'Date', 'crop', 'spacing (m)', 'canopy_avg'])
RTMA_df['soil type'].replace(['sand', 'loamy sand', 'loam'], [0, 1, 2], inplace=True)

display(RTMA_df)
```

```python
# Save target as y data

ydat_RTMA = RTMA_df['target']
display(ydat_RTMA)

# Save everything else as x data

xdat_RTMA = RTMA_df.drop(columns =['target'])
display(xdat_RTMA)
```
